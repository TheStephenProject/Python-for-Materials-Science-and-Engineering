# Bond Energy vs Distance

Atoms in solid materials like steel and aluminum are held together with chemical bonds. Atoms of solid materials are more stable when they are chemically bonded together, and it takes energy to separate atoms which are joined together by a chemical bond. The bonding energy associated with a chemical bond describes the amount of energy needed to separate two atoms from their equilibrium positions to a substantial distance apart. In this section weâ€™ll build a line plot with Python and Matplotlib that describes the bond energy compared to the separation distance of a Na+ Cl- ion pair.

```python
# Importing libaries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Creating a numpy array of r-values

```python
# create an array of inter-atomic distances, r
r = np.arange(0.01,1.01,0.01)
```

## Create arrays for the attractive energy term, the repulsive energy term and the total energy


```python
Ea = -1.436/r
Er = (7.32e-6)/(r**8)
En = Ea + Er
```

## Build a line plot

```python
# Set up the figure size for the plot
plt.figure(figsize=(10, 8))

# Plot the attractive, repulsive, and total energy curves as functions of distance (r)
plt.plot(r, Ea)  # Attractive energy (Ea) vs. distance (r)
plt.plot(r, Er)  # Repulsive energy (Er) vs. distance (r)
plt.plot(r, En)  # Total energy (En) vs. distance (r)

# Set the x-axis limit to focus on the region from 0 to 1 nm
plt.xlim([0, 1])

# Set the y-axis limit to focus on the energy range from -10 to 10 eV
plt.ylim([-10, 10])

# Add a title to the plot
plt.title('Attractive, Repulsive, and Total Energy of an Na+ Cl- Ion Pair')

# Label the x-axis as the distance r in nanometers
plt.xlabel('r (nm)')

# Label the y-axis as bond energy in electron volts (eV)
plt.ylabel('Bond Energy (eV)')

# Add a legend to differentiate between the attractive, repulsive, and total energy curves
plt.legend(["Ea - attractive", "Er - repulsive", "En - total"])

# Display a grid to improve readability of the plot
plt.grid()

# Uncomment the line below to save the plot as a PNG file with a resolution of 72 dpi
# plt.savefig('energy_vs_distance_plot.png', dpi=72)

# Display the plot
plt.show()
```

# Force vs distance

```python
# differentiate the Energy vs. distance curve to create a force vs. distance curve
F = np.diff(En)   # Bond force in eV/nm
F = 1.60217*F     # Bond force in nN
r_F = np.linspace(0.01,1.00,len(F))
```

```python
# create a figure with two plots on it, one plot for the Engergy curve and one plot for the force curve

# create the figure and axis objects
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, figsize=(9,9))

# create the energy vs distance curve
ax1.plot(r,En,'red')
ax1.set_xlim([0.01,1.00])
ax1.set_ylim([-10,10])
ax1.set_title('Bond Energy vs. Distance')
ax1.set_xlabel('Inter-atomic distance (nm)')
ax1.xaxis.set_label_coords(0.5, 0.6)
ax1.set_ylabel('Bond Energy (eV)')
ax1.legend(['En - total energy'])
ax1.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')
ax1.spines['bottom'].set_position(('data', 0.0))

# create the force vs distance curve
ax2.plot(r_F,F)
ax2.set_xlim([0.01, 1.00])
ax2.set_ylim([-0.25, 0.25])
ax2.set_title('Bond Force vs. Distance')
ax2.set_xlabel('Inter-atomic distance (nm)')
ax2.set_ylabel('Bond Force (nN)')
ax2.legend(['F - bond force'])
ax2.spines['top'].set_color('none')
ax2.spines['right'].set_color('none')
ax2.spines['bottom'].set_position(('data', 0.0))

# save and show the plot
plt.savefig("Energy_and_Force_plot.png", dpi=72)
plt.show()
```

# Finding the equalibrium bond energy and the equalibrium bond distance

```python
# calcuate the equalibrium bond energy and the equalibrium bond distance
E_0 = np.min(En)
r_0 = r[En==E_0][0]
print(f"The Equalibrium bond energy is: {round(E_0,2)} eV")
print(f"The Equalibrium bond distance is: {round(r_0,2)} nm")
```
